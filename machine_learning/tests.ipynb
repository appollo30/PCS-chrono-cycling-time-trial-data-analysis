```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

# Riders

```python
df = pd.read_csv("data/riders.csv")
df.head()
```

```python
plt.scatter(df['birth_year'], df['tt'])
plt.xlabel('Birth Year')
plt.ylabel('TT')
plt.title('TT vs Birth Year')
plt.show()
```

```python
# Prepare the data
X = df[['weight']]
y = df['climber']

# Fit linear regression model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Predict values
y_pred = reg.predict(X)

# Plot
plt.scatter(df['weight'], df['climber'],label='Actual')
plt.plot(df['weight'], y_pred, color="orange", label='Regression line')
plt.xlabel('Weight')
plt.ylabel('Climber')
plt.title('Linear Regression: Climber vs Weight')
plt.legend()
plt.show()
```

# Races

```python
races = pd.read_csv("data/races.csv")
print(races.shape)
races.head()
```

```python
#races_important = races[races["race_ranking"] < 100]
races_important = races

races_important.head()
```

```python
sns.pairplot(races_important.select_dtypes(include=[np.number]))
plt.show()
```

```python
races_important.sort_values("winner_speed", ascending=True)
```

```python
races_important = races_important[~(races_important["vertical_meters"].isna() | races_important["startlist_quality"].isna() | races_important["profile_score"].isna())]

X = races_important[["distance", "vertical_meters", "startlist_quality", "profile_score"]]
y = races_important["winner_speed"]
```

```python
# Since the data is sorted in reverse chronological order, we split by index to preserve order
test_size = 0.2  # 20% for test
n_total = len(X)
n_test = int(n_total * test_size)

X_train = X.iloc[n_test:]
X_test = X.iloc[:n_test]
y_train = y.iloc[n_test:]
y_test = y.iloc[:n_test]

print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")
```

```python
# Fit linear regression model on training set
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)

# Predict on test set
y_pred = reg.predict(X_test)

# Build error dataframe for test set
error_df = races_important.loc[X_test.index, ["race_title", "url", "distance", "winner_time"]].copy()
error_df["actual_speed"] = y_test.values
error_df["predicted_speed"] = y_pred
error_df["predicted_time"] = 3600 * error_df["distance"] / error_df["predicted_speed"]
error_df["time_error"] = error_df["winner_time"] - error_df["predicted_time"]
error_df["time_error_abs"] = np.abs(error_df["time_error"])
mse = np.mean((error_df["winner_time"] - error_df["predicted_time"]) ** 2)
median_error = np.median(error_df["time_error_abs"])
print(mse, median_error)
error_df.head()
```
